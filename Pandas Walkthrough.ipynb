```python
# accesing counts of unique items
frog_data.continent.value_counts()
```

```python
# accessing df using conditions
frog_data[frog_data['continent'] != 'Africa'].continent.value_counts()
```

```python
# or bitwise
frog_data[(frog_data['continent'] == 'Africa') | (frog_data['continent'] == 'unknown or invalid')]
```

```python
frog_data.continent.unique()
```

```python
# and bitwise
frog_data[(frog_data['continent'] == 'Central America') & (frog_data['country'] == 'Costa Rica') & (frog_data['ppt_mean'] > 280)]
```

```python
# string manipulation
frog_data[~frog_data.continent.str.contains('Af') &  ~frog_data.country.str.contains('Au')]
```

```python
# accessing column names and values individually
for x,y in frog_data.iteritems():
    num = y.tolist()
    print(num)
```

```python
new_list = []
for nums in num:
    new_num = nums + 200
    new_list.append(new_num)
```

```python
new_list
```

```python
frog_data['country'] = new_list
```

```python
frog_data
```

```python
for x,y in frog_data.iteritems():
    print('Column name: ', x)
    print('-----------------------')
    print('Column values: ', y)
    print('########################')
```

```python
# applying function to a df as a whole
frog_data['fresh_col'] = frog_data.apply(lambda x: x.tmin_mean + 300, axis = 1)
```

```python
# applying function to a specific df column
frog_data['Bad_col'] = frog_data.tmin_mean.apply(lambda x: x + 400)
```

```python
frog_data
```

```python
# creating a new column with a new addition
frog_data['approach 3'] = frog_data['tmin_mean'] + 500
```

```python
frog_data
```

```python
frog_data.columns
```

```python
# dropping a column
frog_data = frog_data.drop(['Unnamed: 0', 'fresh_col', 'Bad_col', 'approach 3'], axis = 1)
```
